#2. Cleaning the data and merging into datasets

```python
driving = driving.rename(columns={"State ":"State"})
driving.head()
#rate of deaths per 100,000.
#rates based on less than 20 deaths are suppressed.
#rates of deaths involved in a accident with an individual with a BAC over 0.08
```

```python
#filtering for years 2012 and 2014.
filtered_drinking = drinking[(drinking["Year"] == 2014)]
combined_drinking = drinking[(drinking["Year"] == 2012) | (drinking["Year"] == 2014)]
combined_drinking.head()
```

```python
merged = pd.merge(driving, filtered_drinking, on="State", how="right")
```

```python
merged_2 = pd.merge(driving, combined_drinking, on="State", how="right")
```

```python
#west virginia was formatted with a "," so needed to remove it to do my whole operation to lat lon pairs.
def fixwestvirg(df):
  df.loc[df["State"]=="West Virginia","Location"] = "West Virginia\n(38.665511, -80.712638)"
  return df

def strip_lat_lon(df):
  lat = df["Location"].str.strip('AaBbCcDdEeFfGgHhIiJjKkLlMmNnOoPpQqRrSsTtUuVvWwXxYyZz\ ').str.split(",").str[0]
  lon = df["Location"].str.strip('AaBbCcDdEeFfGgHhIiJjKkLlMmNnOoPpQqRrSsTtUuVvWwXxYyZz\ ').str.split(",").str[1]
  stripped_lat = lat.str.strip()
  stripped_lon = lon.str.strip()
  df["lat"] = stripped_lat.str.replace("(","")
  df["lon"] = stripped_lon.str.replace(")","")
  df["lat"] = pd.to_numeric(df["lat"])
  df["lon"] = pd.to_numeric(df["lon"])
  return df


def normalizeddeaths(df):
  df["All Ages"] = 0
  df["All Ages"] = np.where(df["Year"] == 2012, df["All Ages, 2012"],
                            np.where(df["Year"] == 2014, df["All Ages, 2014"], df["All Ages"]))

  return df
```

##2 b. Using Scattergeo in the plotly go package before finding the scatter_geo in plotly express.

```python
def hex_to_RGB(hex_str):
    """ #FFFFFF -> [255,255,255]"""
    #Pass 16 to the integer function for change of base
    return [int(hex_str[i:i+2], 16) for i in range(1,6,2)]

def get_color_gradient(c1, c2, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]
```

Here I am making a color gradient for the level of alcohol consumption in each state then assigning based on the sorted index.

```python

merged['text'] = merged['State'] + '<br> Ethanol Consumed in L (per year) ' + (merged['All beverages (Per capita consumption)']).astype(str)+' per capita'

color1 = "#00FFF2"
color2 = "#FF3400"

colors = get_color_gradient(color1, color2, 52)
states = []

merged.sort_values(by=['All beverages (Per capita consumption)'], inplace=True)
index_list = merged.index.tolist()
index_list = [x for x in range(len(index_list))]

#Assign colors to each row based on the index, it is assumed that the length of color and number of rows are same
merged['color']= [colors[x] for x in index_list]
merged["All Ages, 2012"] = merged["All Ages, 2012"].fillna(0)
```

```python
fig = go.Figure(go.Scattergeo(
    locationmode = 'USA-states',
    lon = merged['lon'],
    lat = merged['lat'],
    text = merged['text'],
    marker = dict(
        size = merged["All Ages, 2012"]*100,
        color = merged["color"], # Assign the entire color list
        line_color='rgb(40,40,40)',
        line_width=0.5,
        sizemode = 'area',
    ),
))

fig.update_layout(
        title_text = 'Impaired Driving Deaths per 100,000 <br> With color corresponding to level of drinking (Liters of Ethanol per Year)',
        showlegend = False,
        margin={"r":0,"t":50,"l":0,"b":0},

        geo = dict(
            scope = 'usa',
            landcolor = 'rgb(217, 217, 217)',
        )
    )

fig.show()
import plotly.io as pio
from google.colab import files
pio.write_html(fig, file='HexColorDotMap.html', auto_open=True)
files.download('HexColorDotMap.html')
```

#3. More data manipultaion to get seperate columns for each state for types of alcohol consumption. From this I could use multiple sliders to show the different types of alcohol consumption

```python
#seperate rows for types of alcohol
beer = merged_2[['State',"Year", 'lat',"lon", 'Beer (Per capita consumption)']].copy()
wine = merged_2[['State', "Year",'lat',"lon", 'Wine (Per capita consumption)']].copy()
spirits = merged_2[['State',"Year", 'lat',"lon", 'Spirits (Per capita consumption)']].copy()

beer['Alc Consumption'] = beer['Beer (Per capita consumption)']
beer['type'] = 'beer'
wine['Alc Consumption'] = wine['Wine (Per capita consumption)']
wine['type'] = 'wine'
spirits['Alc Consumption'] = spirits['Spirits (Per capita consumption)']
spirits['type'] = 'spirits'

data= pd.concat([beer, wine, spirits])
```

```python
#Seperate rows for years
year12 = driving[["All Ages, 2012","State"]]
year12["All Ages"] = year12["All Ages, 2012"]
year12["Year"] = 2012
year14 = driving[["All Ages, 2014","State"]]
year14["All Ages"] = year14["All Ages, 2014"]
year14["Year"] = 2014
years = pd.concat([year12, year14])
```

```python
#merging the datasets together to have seperate rows for types of alcohol per year for each state.
final = pd.merge(data, years, on=['State', 'Year'], how='right')
final.fillna(0, inplace=True)
final = final[final["State"]!="United States"]
```
